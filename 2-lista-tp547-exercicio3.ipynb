Você paga 1 real e pode lançar quatro dados. Se a soma dos olhos nos dados for
inferior a 9, recebe de volta r reais, caso contrário perde o investimento de 1 reais. Suponha que r = 10. Você vai, então, a longo prazo, ganhar ou perder dinheiro ao jogar este jogo?

```python
import random

# Função para simular o lançamento de um dado
def lancar_dado():
    return random.randint(1, 6)

# Função para simular o jogo
def simular_jogo(num_simulacoes, r):
    total_ganhos = 0

    for _ in range(num_simulacoes):
        soma_dados = sum(lancar_dado() for _ in range(4))
        if soma_dados < 9:
            total_ganhos += r
        else:
            total_ganhos -= 1

    return total_ganhos / num_simulacoes

# Número de simulações
num_simulacoes = 1000000
# Valor recebido se a soma for inferior a 9
r = 10

# Executando a simulação
resultado_medio = simular_jogo(num_simulacoes, r)

print(f"Resultado médio após {num_simulacoes} simulações: {resultado_medio:.2f} reais")
```

Pelo resultado da simulação, o resultado negativo aponta que o jogador vai perder dinheiro.
