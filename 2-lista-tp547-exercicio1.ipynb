Suponha que haja 40 bolas em um chapéu, das quais 10 são vermelhas, 10 são azuis, 10 são amarelas e 10 são roxas. Qual é a probabilidade de obter duas bolas azuis e duas roxas ao tirar 10 bolas aleatoriamente do chapéu? O que muda no resultado caso a bola seja retirada e não reposta.

Se houver 10 bolas de cada cor em um total de 40 bolas, a probabilidade de tirar uma bola de uma cor específica é de 10 em 40, ou seja, 1/4.

Para calcular a probabilidade de tirar duas bolas azuis e duas roxas em 10 tentativas, sem reposição, usaremos a combinação, pois a ordem das bolas que tiramos não importa.

No caso sem reposição, a probabilidade é calculada usando combinações, pois a quantidade de bolas disponíveis muda após cada retirada.

No caso com reposição, a probabilidade de tirar uma bola de uma cor específica permanece constante (1/4).

```python
import math

# Função para calcular combinações C(n, k)
def comb(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

# Número total de bolas de cada cor
n_A = n_R = 10

# Número total de outras bolas
n_O = 20

# Número total de bolas
n_T = 40

# Número de bolas azuis e roxas que queremos tirar
k_A = k_R = 2

# Número de outras bolas que queremos tirar
k_O = 6

# Número total de bolas que queremos tirar
k_T = 10

# Calculando a probabilidade sem reposição
P_E_sem_reposicao = (comb(n_A, k_A) * comb(n_R, k_R) * comb(n_O, k_O)) / comb(n_T, k_T)

# Calculando a probabilidade com reposição
# A probabilidade de tirar uma bola de uma cor específica é sempre 1/4
P_E_com_reposicao = (1/4)**4 * (3/4)**6

print(f"A probabilidade de tirar duas bolas azuis e duas roxas sem reposição é: {P_E_sem_reposicao:.8f}")
print(f"A probabilidade de tirar duas bolas azuis e duas roxas com reposição é: {P_E_com_reposicao:.8f}")
```
