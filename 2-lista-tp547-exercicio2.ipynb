Faça um programa para estimar a probabilidade de obter pelo menos um dado com
seis ao lançar 5 dados.

A probabilidade de não obter um seis em um único lançamento de um dado é 5/6. Então, a probabilidade de não obter um seis em todos os 5 lançamentos é (5/6)^5. A probabilidade de obter pelo menos um seis é então 1 menos essa probabilidade.

```python
# Probabilidade de não obter um seis em um único lançamento
prob_nao_seis = 5/6

# Probabilidade de não obter um seis em 5 lançamentos
prob_nao_seis_5_lancamentos = prob_nao_seis ** 5

# Probabilidade de obter pelo menos um seis em 5 lançamentos
prob_pelo_menos_um_seis = 1 - prob_nao_seis_5_lancamentos

print(f"A probabilidade de obter pelo menos um seis ao lançar 5 dados é: {prob_pelo_menos_um_seis:.8f}")
```
